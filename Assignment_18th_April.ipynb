```python
#Table shows the life expectancy for an individual born in the United States in  certain years.
#Year of Birth Life Expectancy
#1930 59.7
#1940 62.9
#1950 70.2
#1965 69.7
#1973 71.4
#1982 74.5
#1987 75
#1992 75.7
#2010 78.7
#a. Decide which variable should be the independent variable and which should be  the dependent variable
#b. Draw a scatter plot of the ordered pairs.
#c. Find out the linear regression model.
#d. What will be your life expectancy according to this data?
#e. Plot line plot of predicted values from 1930 to 2010.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
years_of_birth = np.array([1930, 1940, 1950, 1965, 1973, 1982, 1987, 1992, 2010])
life_expectancy = np.array([59.7, 62.9, 70.2, 69.7, 71.4, 74.5, 75.0, 75.7, 78.7])

# Reshape the data for sklearn
X = years_of_birth.reshape(-1, 1)  # Independent variable
y = life_expectancy.reshape(-1, 1) # Dependent variable

```

```python
# Step b: Scatter plot
plt.scatter(years_of_birth, life_expectancy)
plt.xlabel('Year of Birth')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs Year of Birth')
plt.grid(True)
plt.show()
```

```python
# Step c: Linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step d: Predict life expectancy for a given year
year_to_predict = 1985
predicted_life_expectancy = model.predict([[year_to_predict]])
print("Predicted life expectancy in", year_to_predict, ":", predicted_life_expectancy[0][0])
```

```python
# Step e: Plot line plot of predicted values from 1930 to 2010
years_to_predict = np.arange(1930, 2011).reshape(-1, 1)
predicted_life_expectancies = model.predict(years_to_predict)

plt.plot(years_to_predict, predicted_life_expectancies, color='red')
plt.scatter(years_of_birth, life_expectancy)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Predicted Life Expectancy from 1930 to 2010')
plt.grid(True)
plt.show()
```

```python
#The maximum discount value of the Entertainment® card for the “Fine Dining”  section, Edition ten, for various pages is given in Table
#Page number Maximum value ($)
#4 16
#14 19
#25 15
#32 17
#43 19
#57 15
#72 16
#85 15
#90 17
#a. Decide which variable should be the independent variable and which should be the  dependent variable.
#b. Draw a scatter plot of the ordered pairs.
#c. Find the estimated maximum values for the restaurants on page ten and on page 70. d. Plot your predicted values of your model.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
page_number = np.array([4, 14, 25, 32, 43, 57, 72, 85, 90])
max_discount_value = np.array([16, 19, 15, 17, 19, 15, 16, 15, 17])

# Reshape the data for sklearn
X = page_number.reshape(-1, 1)  # Independent variable
y = max_discount_value.reshape(-1, 1) # Dependent variable
```

```python
# Step b: Scatter plot
plt.scatter(page_number, max_discount_value)
plt.xlabel('Page Number')
plt.ylabel('Maximum Discount Value ($)')
plt.title('Maximum Discount Value vs Page Number')
plt.grid(True)
plt.show()
```

```python
# Step c: Linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python

# Predict maximum discount value for page ten and page 70
page_ten = np.array([[10]])
page_seventy = np.array([[70]])
predicted_max_value_page_ten = model.predict(page_ten)
predicted_max_value_page_seventy = model.predict(page_seventy)
print("Estimated maximum value for page ten:", predicted_max_value_page_ten[0][0])
print("Estimated maximum value for page seventy:", predicted_max_value_page_seventy[0][0])
```

```python
# Step d: Plot predicted values
x_values = np.arange(1, 100).reshape(-1, 1)
predicted_values = model.predict(x_values)

plt.plot(x_values, predicted_values, color='red')
plt.scatter(page_number, max_discount_value)
plt.xlabel('Page Number')
plt.ylabel('Maximum Discount Value ($)')
plt.title('Predicted Maximum Discount Value')
plt.grid(True)
plt.show()
```
